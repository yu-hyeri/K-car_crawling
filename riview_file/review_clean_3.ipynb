```python
import pandas as pd
import re
import emoji
from pykospacing import Spacing

# 이모지 제거 함수 정의
def remove_emojis(text):
    if pd.isna(text):  # NaN 값 체크
        return text
    return emoji.replace_emoji(str(text), replace='')

# 특수 문자 제거 함수 정의
def remove_special_characters(text):
    if pd.isna(text):  # NaN 값 체크
        return text
    return re.sub(r'[-+*\^~!.,?♡‥/]', '', str(text))

# 띄어쓰기 교정 함수 정의
spacing = Spacing()

def correct_spacing(text):
    if pd.isna(text):  # NaN 값 체크
        return text
    return spacing(text)

# Excel 파일 읽기
df = pd.read_excel('blog_no_emoji6.xlsx')

# 'title' 열에서 특수문자 제거
df['title'] = df['title'].apply(remove_special_characters)

# 'title' 열의 NaN 값을 빈 문자열로 대체
df['title'] = df['title'].fillna('')

# 'title' 열의 'nan' 문자열을 빈 문자열로 대체
df['title'] = df['title'].replace('nan', '')

# 공백인 행 제거
df['title'] = df['title'].astype(str).str.strip()  # 양쪽 공백 제거
df = df[df['title'].notna() & (df['title'] != '')]  # NaN 값과 빈 문자열 제거

# 'title' 열에서 띄어쓰기 교정
df['title'] = df['title'].apply(correct_spacing)

# 'contents' 열에서 특수문자 제거
df['contents'] = df['contents'].apply(remove_special_characters)

# 'contents' 열의 NaN 값을 빈 문자열로 대체
df['contents'] = df['contents'].fillna('')

# 'contents' 열의 'nan' 문자열을 빈 문자열로 대체
df['contents'] = df['contents'].replace('nan', '')

# 공백인 행 제거
df['contents'] = df['contents'].astype(str).str.strip()  # 양쪽 공백 제거
df = df[df['contents'].notna() & (df['contents'] != '')]  # NaN 값과 빈 문자열 제거

# 'contents' 열에서 띄어쓰기 교정
df['contents'] = df['contents'].apply(correct_spacing)

# 결과 출력
print(df)

# CSV 파일로 저장
df.to_csv('please.csv', index=False)

# XLSX 파일로 저장
df.to_excel('please.xlsx', index=False, engine='openpyxl')
```

```python
import pandas as pd
import re
import emoji
from pykospacing import Spacing

# 이모지 제거 함수 정의
def remove_emojis(text):
    if pd.isna(text):  # NaN 값 체크
        return text
    return emoji.replace_emoji(str(text), replace='')

# 특수 문자와 의미 없는 표현 제거 함수 정의
def remove_special_characters(text):
    if pd.isna(text):  # NaN 값 체크
        return text
    # 특수 문자 제거
    text = re.sub(r'[-+\^~!.,?♡‥/]', '', str(text))
    # 의미 없는 표현 제거 (ㅋ, ㅎ, ㅜ, ㅠ가 2번 이상 반복되는 경우)
    text = re.sub(r'[ㅋㅎㅜㅠㅇㅏ]{2,}', '', text)
    return text

# 띄어쓰기 교정 함수 정의
spacing = Spacing()

def correct_spacing(text):
    if pd.isna(text):  # NaN 값 체크
        return text
    # 텍스트를 문장 단위로 나누어 처리
    sentences = re.split(r'(?<=[.!?])\s+', text)
    corrected_sentences = []
    for sentence in sentences:
        if len(sentence) > 100:  # 문장이 너무 길 경우 100자 단위로 나누어 처리
            parts = [sentence[i:i+100] for i in range(0, len(sentence), 100)]
            corrected_parts = [spacing(part) for part in parts]
            corrected_sentence = ''.join(corrected_parts)
        else:
            corrected_sentence = spacing(sentence)
        corrected_sentences.append(corrected_sentence)
    return ' '.join(corrected_sentences)

# Excel 파일 읽기
df = pd.read_excel('blog_no_emoji6.xlsx')

# 'title' 열 처리
df['title'] = df['title'].apply(remove_special_characters)
df['title'] = df['title'].fillna('').replace('nan', '')
df['title'] = df['title'].astype(str).str.strip()
df = df[df['title'].notna() & (df['title'] != '')]
df['title'] = df['title'].apply(correct_spacing)

# 'contents' 열 처리
df['contents'] = df['contents'].apply(remove_special_characters)
df['contents'] = df['contents'].fillna('').replace('nan', '')
df['contents'] = df['contents'].astype(str).str.strip()
df = df[df['contents'].notna() & (df['contents'] != '')]
df['contents'] = df['contents'].apply(correct_spacing)

# 결과 출력
print(df)

# CSV 파일로 저장
df.to_csv('please2.csv', index=False)

# XLSX 파일로 저장
df.to_excel('please2', index=False, engine='openpyxl')
```
