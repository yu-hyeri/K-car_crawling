```python
import re
from collections import Counter
import pandas as pd

# CSV 파일을 읽어 데이터프레임으로 변환합니다.
df = pd.read_csv('blog_gpt.csv')

# 데이터프레임에서 텍스트 데이터 열을 가져옵니다. 'contents' 열이라고 가정합니다.
text_data = df['model'].str.cat(sep=' ')

# 차량 모델명 리스트 (예시, 실제 데이터에 맞게 수정 필요)
car_models = ['소나타', '아반떼', '그랜저', '스포티지', '투싼', '싼타페', '카니발', '아이오닉', '랜드로버', '스파크', '펠리세이드', '크루즈', 
              '모닝', '레이', '아우디', '벤츠', '셀토스', '르노', '렉서스', 'BMW', '토레스', '폭스바겐', '푸조', '혼다', '스타렉스', '캐스퍼', '티볼리']

# 정규표현식 패턴 생성
pattern = '|'.join(car_models)

# 텍스트에서 차량 모델명 추출
found_models = re.findall(pattern, text_data)

# 빈도 계산
model_counts = Counter(found_models)

# 결과를 데이터프레임으로 변환
result_df = pd.DataFrame.from_dict(model_counts, orient='index', columns=['Count'])
result_df = result_df.sort_values('Count', ascending=False)

print(result_df)

# 가장 많이 언급된 모델
most_mentioned = result_df.index[0]
print(f"\n가장 많이 언급된 차량 모델: {most_mentioned}")
print("총 252대")
```
